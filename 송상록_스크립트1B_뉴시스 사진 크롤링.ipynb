```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
import re
import os
import pickle
from tqdm import tqdm
import urllib
from urllib.parse import urlparse
```

```python
driver = webdriver.Chrome()
driver.get(url="https://www.newsis.com")
driver.find_element(by=By.XPATH, value='//*[@id="header"]/div[2]/div[2]/div/button/img').click()
search_bar = driver.find_element(by=By.XPATH, value='//*[@id="search_inputbox"]')
search_bar.send_keys("이창용 총재")
driver.find_element(by=By.XPATH, value='//*[@id="search_form_main"]/button/img').click()
driver.find_element(by=By.XPATH, value='//*[@id="header"]/div/div[2]/div/ul/li[3]/a').click()
driver.find_element(by=By.XPATH, value='//*[@id="content"]/div[2]/div[2]/dl[2]/dd/ul/li[3]/a').click()
driver.find_element(by=By.XPATH, value='//*[@id="content"]/div[2]/div[2]/dl[4]/dd[1]/ul/li[6]/a').click()
left_day = driver.find_element(by=By.XPATH, value='//*[@id="detailSdate1"]')
left_day.send_keys("20220422")
right_day = driver.find_element(by=By.XPATH, value='//*[@id="detailEdate2"]')
right_day.send_keys("20240408")
driver.find_element(by=By.XPATH, value='//*[@id="term"]/button').click()

link_storage = dict()

# 2를 40으로 바꿔주세요.
for step in tqdm(range(1, 40)):
    page = driver.find_element(by=By.CSS_SELECTOR, value="div#paging_t1 span.num.scpage a strong").text
    print(f"page {page}")
    news_list = driver.find_elements(by=By.CSS_SELECTOR, value="ul.photoList li")
    for n in news_list:
        pic_link = n.find_element(by=By.CSS_SELECTOR, value="div.thumCont a img").get_attribute("src")
        pic_date = n.find_element(by=By.CSS_SELECTOR, value="div.txtCont p.time").text
        date = re.findall(r"([0-9]{4})\.([0-9]{1,2})\.([0-9]{1,2})", pic_date)
        if date:
            year = date[0][0]
            month = date[0][1]
            day = date[0][2]
            try:
                extension = os.path.splitext(urlparse(pic_link).path)[1]
            except:
                extension = ".jpg"
            title = f"lcy_pictures/ns_{year:0>4}{month:0>2}{day:0>2}_1{extension}"
            i = 1
            while title in link_storage.keys():
                i += 1
                title = f"lcy_pictures/ns_{year:0>4}{month:0>2}{day:0>2}_{i}{extension}"
            link_storage[title] = pic_link 
    if step == 39:
        driver.quit()
    elif step % 10 == 0:
        driver.find_element(by=By.XPATH, value='//*[@id="paging_t1"]/button[3]').click()
        driver.find_element(by=By.XPATH, value='//*[@id="paging_t1"]/span[1]/a').click()
    else:
        next = (step % 10) + 1
        driver.find_element(by=By.XPATH, value=f'//*[@id="paging_t1"]/span[{next}]/a').click()

with open('links_newsis.pkl', 'wb') as f:
    pickle.dump(link_storage, f)
    
print(link_storage)


```
