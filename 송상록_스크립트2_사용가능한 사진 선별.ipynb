```python
from deepface import DeepFace
from retinaface import RetinaFace
import os
import shutil
```

```python
os.chdir('/content/drive/MyDrive/window_dragon')
print(os.listdir('window_dragon_crawling'))
```

* crawling_b1: 2명 이상이 등장한 사진은 제외합니다.
* crawling_b2: 마스크를 썼거나, 이창용 총재 말고 다른 사람이 있는 사진은 제외합니다.
이 사진들은 실제로 사용하진 않습니다. 다만 백업 용도로 남겨둡니다.

```python
from tqdm.auto import tqdm

for image in tqdm(os.listdir('window_dragon_crawling')):
  try:
    full_location = f'window_dragon_crawling/{image}'
    try:
      result = len(RetinaFace.extract_faces(full_location))
      if result != 1:
        print(f'{image} {result} move')
        shutil.move(full_location, 'crawling_b1')
      else:
        print(f'{image} {result} stay')
    except Exception as e:
      print(f"{image} {e}")
      shutil.move(full_location, 'crawling_b2')
  except:
    pass
```
